```python
file_list = ['file_small.conll','file_medium.conll']
file_small = open(file_list[0], 'w')
file_medium = open(file_list[1], 'w')

with open("/content/file_augmented_mod.conll", "r") as file:
  cnt=0
  in_episode = False
  for line in file:
    if line.startswith('#begin'):
      if (cnt<16):
        file_medium.write(line)
        if (cnt<8):
          file_small.write(line)
        in_episode = True

    elif line.startswith('#end'):
      if (cnt<16):
        file_medium.write(line)
        if (cnt<8):
          file_small.write(line)
        in_episode = False
      cnt+=1
      if (cnt == 24):
        cnt = 0
        print(cnt)
        continue
    elif in_episode:
      file_medium.write(line)
      if (cnt < 8):
        file_small.write(line)

print(cnt)
file_small.close()
file_medium.close()
```

```python
for i in range(5):
    with open("/content/file_small.conll", "r") as file:
      in_episode = False
      num = 0
      for line in file:
      #print(line)
        if line.startswith('#begin'):
          if num in random_episodes_small['train'][i]:
            file_new_small = open(files['small'][i][0], 'a')
          else:
            file_new_small = open(files['small'][i][1], 'a')
          file_new_small.write(line)
          in_episode = True

        elif line.startswith('#end'):
          file_new_small.write(line)
          file_new_small.close()
          in_episode = False
          num += 1
          
        elif in_episode:
          file_new_small.write(line)
```

```python
for i in range(5):
    with open("file_medium.conll", "r") as file:
      in_episode = False
      num = 0
      for line in file:
      #print(line)
        if line.startswith('#begin'):
          season = int(line.split()[2][11:13])-1
        
          ep_num = (int(line.split()[2][14:16])-1) + season * 24
          in_episode = True

          if ep_num in random_episodes_medium['train'][i]:
            file_new = open(files['medium'][i][0], 'a')
          else:
            file_new = open(files['medium'][i][1], 'a')
          file_new.write(line)
        elif line.startswith('#end'):
          file_new.write(line)
          file_new.close()
        elif in_episode:
          file_new.write(line)
```
